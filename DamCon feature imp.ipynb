```python
#Get features

# data_feat = pd.read_csv('/Users/ishaku/Documents/PhD/phd_work/csvs/features.csv', index_col=0)
# data_feat = pd.read_csv('/Users/ishaku/Documents/PhD/phd_work/csvs/features_coll.csv', index_col=0)

 ###select features
# data = data_feat[['instance', 'hate_count', 'url_count', 'mentions_count',
#    'hashtag_count', 'hate_avg', 'url_avg', 'mentions_avg', 'hashtag_avg',
#    'users', 'posts', 'nsfw', 'media_removal', 'fed_time_rem',
#     'followers_only',
#    'hate_perc', 'url_perc', 'mentions_perc', 'hash_perc',
#    'followers_count', 'following_count', 'reblogs_count', 
#    'active_halfyear', 'active_month', 'hash_ftr', 'hash_rej', 'hash_sen',
#    'users_tr', 'posts_tr', 'hate_tr', 'url_tr', 'blocked']]

    
data = data.drop(columns=['instance'])

    
  

 #train data
features = data.drop('blocked', axis=1)
labels = data['blocked']

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
data

# check splits
for dataset in [y_train, y_test]:
    print(round(len(dataset) / len(labels), 2))

#save to csv

X_train.to_csv('/Users/ishaku/Documents/PhD/phd_work/algorithm_collateral/train_features.csv', index=False)
X_test.to_csv('/Users/ishaku/Documents/PhD/phd_work/algorithm_collateral/test_features.csv', index=False)

y_train.to_csv('/Users/ishaku/Documents/PhD/phd_work/algorithm_collateral/train_labels.csv', index=False)
y_test.to_csv('/Users/ishaku/Documents/PhD/phd_work/algorithm_collateral/test_labels.csv', index=False)


#Load data

tr_features = pd.read_csv('/Users/ishaku/Documents/PhD/phd_work/algorithm_collateral/train_features.csv')
tr_labels = pd.read_csv('/Users/ishaku/Documents/PhD/phd_work/algorithm_collateral/train_labels.csv')
   
#Random forest
print(RandomForestClassifier())
print(RandomForestRegressor())


def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))


rf = RandomForestClassifier()
parameters = {
    'n_estimators': [5, 50, 250],
    'max_depth': [2, 4, 8, 16, 32, None]
}

cv = GridSearchCV(rf, parameters, cv=5)
cv.fit(tr_features, tr_labels.values.ravel())

print_results(cv)

# Write pickled model
# joblib.dump(cv.best_estimator_, '/Users/ishaku/Documents/PhD/phd_work/algorithm_collateral/RF_model.pkl')


```

```python
# Generate feature importance plot
from matplotlib import pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 1, figsize=(14,8))

feat_imp = cv.best_estimator_.feature_importances_
# feat_imp = cv.gb_best_estimator.feature_importances_
indices = np.argsort(feat_imp)
plt.yticks(range(len(indices)), [tr_features.columns[i] for i in indices])
plt.barh(range(len(indices)), feat_imp[indices], color='#1f77b4', align='center')
ax.set_facecolor("white")

plt.ylabel('Features', fontsize=20, fontweight="bold")
plt.xlabel('Importance', fontsize=20, fontweight="bold")
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 15)

# ax.set_title('Gradient Boosted Trees',fontsize= 25)
plt.tight_layout()

plt.savefig("/Users/ishaku/Documents/PhD/phd_work/plots/feature_importance_RF_damcon.pdf")
```
